# Axelrod Data Gathering
## Runs a tournament with most strategies in axelrod library, sends results of all players playing against each other 

```python
import axelrod as axl

# Not included AMM
axl_demo = [s() for s in axl.demo_strategies]
tournament = axl.Tournament(axl_demo)
results = tournament.play()
results.write_summary("Axl2.csv")

plot = axl.Plot(results)
pPd = plot.pdplot()
```
