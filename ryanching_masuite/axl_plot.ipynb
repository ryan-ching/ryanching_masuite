```python
### Using Liems plotting code for Axelrod ###
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
import numpy as np
import csv

import numpy as np
import pandas as pd
from pandas import DataFrame as df
import axelrod as axl
```

```python
def axl_df(data_name):
    data = pd.read_csv(data_name)
    axl_df = df({'Index': [], 'Rank': [], 'Name': [], 'Median Score': [], 
                 'Cooperation Rating': [], 'Wins': [], 'Initial C Rate': [], 'CC Rate': [], 
                 'CD Rate': [], 'DC Rate': [], 'DD Rate': [], 'CC to C Rate': [],
                 'CD to C Rate': [], 'DC to C Rate': [], 'DD to C Rate': []})
    for idx in np.arange(0,len(data.index)):
        ranks = data.at[idx, 'Rank']
        names = data.at[idx, 'Name']
        median_score = data.at[idx, 'Median_score']
        coop_rating = data.at[idx, 'Cooperation_rating']
        wins = data.at[idx, 'Wins']
        initial_c_rate = data.at[idx, 'Initial_C_rate']
        cc_rate = data.at[idx, 'CC_rate']
        cd_rate = data.at[idx, 'CD_rate']
        dc_rate = data.at[idx, 'DC_rate']
        dd_rate = data.at[idx, 'DD_rate']
        cc_to_c_rate = data.at[idx, 'CC_to_C_rate']
        cd_to_c_rate = data.at[idx, 'CD_to_C_rate']
        dc_to_c_rate = data.at[idx, 'DC_to_C_rate']
        dd_to_c_rate = data.at[idx, 'DD_to_C_rate']
        axl_df.loc[len(axl_df.index)] = [idx, ranks, names, median_score, coop_rating, 
                                         wins, initial_c_rate, cc_rate, cd_rate, dc_rate, 
                                         dd_rate, cc_to_c_rate, cd_to_c_rate, dc_to_c_rate,
                                         dd_to_c_rate] 
    return axl_df
```

```python
axl_file_name = "Axl_All.csv"
axl_data = axl_df(axl_file_name)

axl_ranks = axl_data['Rank']
axl_med_score = axl_data['Median Score']
axl_coop_rating = axl_data['Cooperation Rating']
axl_wins = axl_data['Wins']
axl_initial_c_rate = axl_data['Initial C Rate']
axl_cc_rate = axl_data['CC Rate']
axl_cd_rate = axl_data['CD Rate']
axl_dc_rate = axl_data['DC Rate']
axl_dd_rate = axl_data['DD Rate']
axl_cc_to_c_rate = axl_data['CC to C Rate']
axl_cd_to_c_rate = axl_data['CD to C Rate']
axl_dc_to_c_rate = axl_data['DC to C Rate']
axl_dd_to_c_rate = axl_data['DD to C Rate']
```
